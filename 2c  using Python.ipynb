```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv("weather_data HCM 2c.csv")
df.head()
```

```python
x, y = np.array(df["humidity"]).reshape((-1,1)),np.array(df["Rain"])
#  kiểm tra thông tin chuỗi x,y
print(x)
print(y)
```

```python
model = LogisticRegression(solver="liblinear", C=10.0, random_state = 0)
model.fit(x,y)
```

lấy giá trị B0

```python
model.intercept_
```
