```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
db = pd.read_csv('Colleges-and-Universities.csv')
```

```python
db
```

```python
x, y = np.array(db[["Acceptance Rate","Expenditures/Student", 'Median SAT']]).reshape((-1, 3)), np.array( db["Graduation %"]).reshape((-1, 1)) 
```

```python
model = LinearRegression()
```

```python
model.fit(x,y)  
```

```python
print("Hệ số chắn:", model.intercept_) 
print("Hệ số phụ thuộc:", model.coef_) 
print("Điểm R Square:", model.score(x, y))
```
